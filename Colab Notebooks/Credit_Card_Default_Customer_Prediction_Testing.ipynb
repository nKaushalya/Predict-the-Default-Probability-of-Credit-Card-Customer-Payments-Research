```python
# Loading the dataset
credit_card = pd.read_csv('/content/drive/My Drive/w1790011_FYP_6COSC023W.Y2023/Data Set/credit_card_default_payments_cleaned_data_set.csv')
print(credit_card.columns)
```

```python
#Load the Model File
import pickle
model_x = pickle.load(open('/content/drive/My Drive/w1790011_FYP_6COSC023W.Y2023/Prediction Model/credit_card_finalized_model.sav', 'rb'))
```

```python
#Test the model using sample of default customer
predictionProbability = model_x.predict_proba([[82, 39, 130000, 1590000,1, 2, 119047, 0, 0, 0, 3500, 4000, 3136, 1843, 1885, 1848]])
prediction = model_x.predict([[82, 39, 130000, 1590000,1, 2, 119047, 0, 0, 0, 3500, 4000, 3136, 1843, 1885, 1848 ]])
print('Probability:', predictionProbability)
print('Probability %:', predictionProbability*100)
print('Prediction:', prediction)
```

```python
#Test the model using sample of non-default customer
predictionProbability = model_x.predict_proba([[108, 40, 128000, 4425000, 2, 2, 127360, 7351, 0, 0, 260, 20000, 0, 43, 0, 8]])
prediction = model_x.predict([[108, 40, 128000, 4425000, 2, 2, 127360, 7351, 0, 0, 260, 20000, 0, 43, 0, 8]])
print('Probability:', predictionProbability)
print('Probability %:', predictionProbability*100)
print('Prediction:', prediction)
```

```python
#Test the model using large sample of customer data
sampleData = credit_card[:1000]
sampleDataFeatures = np.asarray(sampleData.drop(['Payment Default in Month N+1', 'Unnamed: 0', 'Nature of Business','Credit Segment','Method Month N-1','Method Month N','Application Score','Product','Method Month N-2','Employment Status','Gender','Retail Segment'], axis=1))
print(sampleDataFeatures)

predictionProbability = model_x.predict_proba(sampleDataFeatures)
prediction = model_x.predict(sampleDataFeatures)
print('Probability:', predictionProbability)
print('Probability %:', predictionProbability*100)
print('Prediction:', prediction)
```
